```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.data()
```

```python
"when we import the data, thier all info is in the data,and others"
"info is in the feature_names and more "
df = pd.DataFrame(iris.data,columns = iris.feature_names)
df
```

```python
"this will create target column and directly add the column in the dataset"
df["flower"] = iris.target
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df[0:50]
```

```python
"we have to saperate the unique outputs"
df0 = df[0:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
df0.head()
```

```python
"here we plot a scatter plot of sepal length and width of three types"
plt.scatter(df0["sepal length (cm)"],df0["sepal width (cm)"],label = "setosa")
plt.scatter(df1["sepal length (cm)"],df1["sepal width (cm)"],label = "versicolor")
plt.scatter(df2["sepal length (cm)"],df2["sepal width (cm)"],label = "virginica")
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("sepal distribution")
plt.legend()
plt.show()
```

```python
"here we plot a scatter plot of petal length and width of three types"
plt.scatter(df0["petal length (cm)"],df0["petal width (cm)"])
plt.scatter(df1["petal length (cm)"],df1["petal width (cm)"])
plt.scatter(df2["petal length (cm)"],df2["petal width (cm)"])
plt.show()
```

```python
X =df.drop("flower" , axis = 1)
y = df.flower
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(X_train , y_train)
model.score(X_test, y_test)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[4.5,3.5,1.5,0.4]])
```

```python
"we use this for loop for, at which point the accuracy remain constant"
score=[]

for i in range(1,15):
    model = KNeighborsClassifier()
    model.fit(X_train , y_train)
    accuracy = model.score(X_test, y_test)
    score.append(accuracy)
score
```

```python
plt.plot(range(1,15),score)
plt.show()
```

```python
"we have  stored the new values in predicted "
predicted = model.predict(X_test)
predicted
```

```python
"these are already predicted values"
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,predicted)
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm , annot = True)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predicted))
```
